```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df.csv')
df
```

```python
import random
```

```python
scrambled_commments = [i.split(' ') for i in df['processed_text']]

for i in scrambled_commments:
    random.shuffle(i)
```

```python
scrambled_commments
```

```python
# Shuffle the list in the lemmatized_text column
df['lemmatized_text'] = scrambled_commments
```

```python
df.processed_text = df.lemmatized_text.apply(lambda x: ' '.join(x))
```

```python
df
```

```python
# drop the text column
df.drop('text', axis=1, inplace=True)
```

```python
df
```

```python
# drop the datetime column
df.drop('datetime', axis=1, inplace=True)
```

```python
# drop the score column
df.drop('score', axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv('df_anon.csv', index=False)
```
